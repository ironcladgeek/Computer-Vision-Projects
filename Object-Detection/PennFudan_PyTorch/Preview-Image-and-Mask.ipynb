```python
import os
import numpy as np
import torch
from PIL import Image
import matplotlib.pyplot as plt
```

```python
base_dir = os.path.abspath('../../_data/PennFudanPed')
images_dir = os.path.join(base_dir, 'PNGImages')
masks_dir = os.path.join(base_dir, 'PedMasks')
```

```python
imgs = list(sorted(os.listdir(images_dir)))
masks = list(sorted(os.listdir(masks_dir)))
```

```python
imgs[:2], masks[:2]
```

```python
idx = np.random.choice(len(imgs))
img = Image.open(os.path.join(images_dir, imgs[idx])).convert('RGB')
mask = np.array(Image.open(os.path.join(masks_dir, masks[idx])))
```

```python
obj_ids = np.unique(mask)[1:]
ms = mask == obj_ids[:, None, None]
```
